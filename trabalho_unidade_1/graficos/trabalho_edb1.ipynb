```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
selec = pd.read_csv('selectionsort.csv')
```

```python
valores = selec["n"]
ultimo_tempo = selec["tempo_ms"].iloc[-1]

base_nquad = valores ** 2
base_n = valores
base_nlogn = valores * np.log2(valores)
base_logn = np.log2(valores)

# Fórmula: (ultimo_tempo_empirico / ultimo_valor_base) * base
selec["Teorico_O(n^2)"] = (ultimo_tempo / base_nquad.iloc[-1]) * base_nquad
selec["Teorico_O(n)"] = (ultimo_tempo / base_n.iloc[-1]) * base_n
selec["Teorico_O(n log n)"] = (ultimo_tempo / base_nlogn.iloc[-1]) * base_nlogn
selec["Teorico_O(log n)"] = (ultimo_tempo / base_logn.iloc[-1]) * base_logn

plt.figure(figsize=(10, 6))

plt.plot(selec["n"], selec["tempo_ms"], label="Empírico (Selection Sort)", marker='o', color='black', linewidth=2)

plt.plot(selec["n"], selec["Teorico_O(n^2)"], label="Teórico O(n²)", linestyle='--', color='red')
plt.plot(selec["n"], selec["Teorico_O(n log n)"], label="Teórico O(n log n)", linestyle='--', color='blue')
plt.plot(selec["n"], selec["Teorico_O(n)"], label="Teórico O(n)", linestyle='--', color='green')
plt.plot(selec["n"], selec["Teorico_O(log n)"], label="Teórico O(log n)", linestyle='--', color='purple')

# Formatação
plt.xlabel("Quantidade de elementos (n)")
plt.ylabel("Tempo de execução (ms)")
plt.legend(loc="upper left")
plt.grid(True)
```
